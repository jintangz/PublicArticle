```python
import numpy as np
import matplotlib.pyplot as plt

#设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#绘制服从0到1均匀分布的概率密度函数曲线和累计分布函数曲线
x = np.linspace(0, 1, 100)
pdf = np.ones_like(x)
cdf = x
fig = plt.figure(figsize=(8, 4))
ax=fig.add_subplot(1, 2, 1)
ax.plot(x, pdf, label='概率密度函数', color='blue')
x1=np.linspace(-1,0,100)
ax.plot(x1, np.zeros_like(x1), color='blue')
x2=np.linspace(1,2,100)
ax.plot(x2, np.zeros_like(x2), color='blue')
ax.set_title('f(x)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax=fig.add_subplot(1, 2, 2)
ax.plot(x, cdf, label='累计分布函数', color='orange')
x3=np.linspace(1,2,100)
ax.plot(x3, np.ones_like(x2), color='orange')
x4=np.linspace(-1,0,100)
ax.plot(x4, np.zeros_like(x4), color='orange')
ax.set_title('F(x)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.suptitle('U(0,1)的PDF和CDF', fontsize=16)
```
