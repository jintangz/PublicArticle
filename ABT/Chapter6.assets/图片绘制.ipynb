```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
# 绘制AB实验的学习效应示意图
def plot_learning_effect():
    # 数据
    x = np.arange(1, 11)  # 1到10的学习周期
    y_A = np.log(np.log(np.array(x))+1) # A组的学习效果

    # 绘图
    # 创建ax对象
     
    fig, axs = plt.subplots(1, 2, figsize=( 10,5), sharey=True,sharex=True)  # 创建1行2列的子图，设置共享y轴
    ax1 = axs[0]
    ax1.plot(x, y_A, marker='o',  color='orange', label= '整体')

    
    # 添加标题和标签
    # ax1.set_title('学习效应')
    ax1.set_xlabel('实验周期(周)')
    # 隐藏y轴刻度
    ax1.tick_params(axis='y', which='both', left=True, right=False, labelleft=False, labelright=False)
    ax1.set_ylabel('实验效应')
    ax1.legend().set_frame_on(False)  # 隐藏图例边框
    ax1.legend().set_loc((0.9, 0.8))  #
    
    ax1.set_xlim(0, 11)  # 设置x轴范围
    # 添加网格和图例
    # ax1.legend().set_frame_on(False)  # 隐藏图例边框
    # ax1.legend().set_loc((0.9, 0.6))  # 图例位置
    
    ax1.spines['top'].set_visible(False)  # 隐藏上边框
    ax1.spines['right'].set_visible(False)  # 隐藏

    ax2 = axs[1]
    #绘制新老用户的实验效应
    y_B = np.log(np.log(np.array(x))+1)  # B组的学习效果
    # 绘制新用户效应 在0.4处上下波动
    y_new = 0.4 + 0.05 * np.random.randn(10)  # 新用户的实验效应
    ax2.plot(x, y_new, marker='o', label='新用户', color='gold')
    ax2.plot(x, y_B, marker='o', label='老用户', color='dodgerblue')
    # ax2.set_title('新老用户的实验效应')
    ax2.set_xlabel('实验周期(周)')
    ax2.legend().set_frame_on(False)  # 隐藏图例边框
    ax2.legend().set_loc((0.9, 0.75))  # 图例位置
    ax2.spines['top'].set_visible(False)  # 隐藏上边框
    ax2.spines['right'].set_visible(False)  # 隐藏右边框
    ax2.set_xlim(0, 11)  # 设置x轴范围
    # 显示图形
    plt.suptitle('学习效应示意', fontsize=16, fontweight='bold')  # 设置总标题
    plt.show()
plot_learning_effect()
```

```python
# 绘制AB实验的学习效应示意图
def plot_curiosity_effect():
    # 数据
    x = np.array(range(10, 0, -1))  # 1到10的学习周期
    y_A = 1/np.log((np.log(np.array(x))+1)) # A组的学习效果

    # 绘图
    # 创建ax对象
     
    fig, axs = plt.subplots(1, 2, figsize=( 10,5), sharey=True,sharex=True)  # 创建1行2列的子图，设置共享y轴
    ax1 = axs[0]
    ax1.plot(x, y_A, marker='o',  color='orange', label= '整体')

    
    # 添加标题和标签
    # ax1.set_title('学习效应')
    ax1.set_xlabel('实验周期(周)')
    # 隐藏y轴刻度
    ax1.tick_params(axis='y', which='both', left=True, right=False, labelleft=False, labelright=False)
    ax1.set_ylabel('实验效应')
    ax1.legend().set_frame_on(False)  # 隐藏图例边框
    ax1.legend().set_loc((0.9, 0.8))  #
    
    ax1.set_xlim(0, 11)  # 设置x轴范围
    # 添加网格和图例
    # ax1.legend().set_frame_on(False)  # 隐藏图例边框
    # ax1.legend().set_loc((0.9, 0.6))  # 图例位置
    
    ax1.spines['top'].set_visible(False)  # 隐藏上边框
    ax1.spines['right'].set_visible(False)  # 隐藏

    ax2 = axs[1]
    #绘制新老用户的实验效应
    y_B = 1/np.log((np.log(np.array(x))+1))  # B组的学习效果
    # 绘制新用户效应 在0.4处上下波动
    y_new = 1.0 + 0.05 * np.random.randn(10)  # 新用户的实验效应
    ax2.plot(x, y_new, marker='o', label='新用户', color='gold')
    ax2.plot(x, y_B, marker='o', label='老用户', color='dodgerblue')
    # ax2.set_title('新老用户的实验效应')
    ax2.set_xlabel('实验周期(周)')
    ax2.legend().set_frame_on(False)  # 隐藏图例边框
    ax2.legend().set_loc((0.9, 0.75))  # 图例位置
    ax2.spines['top'].set_visible(False)  # 隐藏上边框
    ax2.spines['right'].set_visible(False)  # 隐藏右边框
    ax2.set_xlim(0, 11)  # 设置x轴范围
    # 显示图形
    plt.suptitle('新奇效应示意', fontsize=16, fontweight='bold')  # 设置总标题
    plt.show()
plot_curiosity_effect()
```

```python
# subplot2grid 绘制多子图，一个子图可能占多个子图区域

# --- IGNORE ---
import matplotlib.gridspec as gridspec
def plot_subplot2grid_example():
    fig = plt.figure(figsize=(10, 6))  # 创建一个图形对象
    gs = gridspec.GridSpec(2, 2)  # 创建3x3的网格布局
    
    #设置子图共享 X轴 和 Y轴


    # 绘制第一个子图，占据第一行的前两列
    ax1 = fig.add_subplot(gs[0, :1])
    ax1.set_title('早晚高峰效应')
    # 绘制早晚高峰时序图，图形应在早晚高峰时段有较明显的峰值

    x = np.arange(0, 24, 1)
    #模拟一天的变化
    y = np.random.random(24) * 5  # 模拟一天的变化
    y[7:10] += np.random.random(3)*40  # 模拟早高峰
    y[17:20] += np.random.random(3)*50  # 模拟晚高峰
    ax1.plot(x, y, marker='o', label='早晚高峰时序图', color='dodgerblue')
    ax1.set_xlabel('时间 (小时)')
    ax1.set_ylabel('流量')
    ax1.tick_params(axis='x', which='both', bottom=True, top=False, labelbottom=False, labeltop=False)

    # 绘制第二个子图，占据第一行的最后一列
    ax2 = fig.add_subplot(gs[0, 1])
    ax2.set_title('周末效应')
    
    # 绘制周末效应图，图形应在周末时段有较明显的峰值
    x_weekend = np.arange(0, 22, 1)
    y_weekend = np.random.random(22) * 5  # 模拟一周的变化
    y_weekend[5:7] += np.random.random(2)*30  # 模拟周末
    y_weekend[12:14] += np.random.random(2)*40  # 模拟周末
    y_weekend[19:21] += np.random.random(2)*40  # 模拟周末
    ax2.plot(x_weekend, y_weekend, marker='o', label='周末效应图', color='orange')
    ax2.set_xlabel('时间 (天)')
    ax2.tick_params(axis='y', which='both', left=True, right=False, labelleft=False, labelright=False)
    ax2.tick_params(axis='x', which='both', bottom=True, top=False, labelbottom=False, labeltop=False)
    # 绘制第三个子图，占据第二行的所有列
    ax3 = fig.add_subplot(gs[1, 0])
    ax3.set_title('实验效应')
    # 绘制实验效应图，图形应在实验周期内有较明显的变化
    y = np.random.random(24)   # 模拟一天的变化
    y[7:10] += np.random.random(3)*2  # 模拟早高峰
    y[17:20] += np.random.random(3)*2  # 模拟晚高峰
    ax3.plot(x, y, marker='o', label='实验效应图', color='blue')
    ax3.set_ylabel('实验效应')
    ax3.set_xlabel('时间(小时)')

    # 绘制第四个子图，占据第二行的所有列
    ax3 = fig.add_subplot(gs[1, 1])
    ax3.set_title('实验效应')
    # 绘制实验效应图，图形应在实验周期内有较明显的变化
    y = np.random.random(22)   # 模拟一天的变化
    y[5:7] += np.random.random(2)*3  # 模拟早高峰
    y[12:14] += np.random.random(2)*3  # 模拟晚高峰
    y[19:21] += np.random.random(2)*3  # 模拟晚高峰
    ax3.plot(x_weekend, y, marker='o', label='实验效应图', color='gold')
    
    ax3.set_xlabel('时间(天)')
    ax3.tick_params(axis='y', which='both', left=True, right=False, labelleft=False, labelright=False)
    plt.tight_layout()
    plt.show()
plot_subplot2grid_example()
```

```python
# 绘制两总体均值的显著性检验
def plot_two_means_significance():
    # 数据
    import scipy.stats as stats
    
    
    x1 = np.linspace(-3, 3, 100)  # A组 
    y1 = stats.norm.pdf(x1, loc=0, scale=1)  # A组正态分布
    x2 = np.linspace(0, 6, 100)  # B组
    y2 = stats.norm.pdf(x2, loc=3, scale=1)  # B组正态分布
    # 绘图
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(x1, y1, label='假设分布', color='dodgerblue')
    ax.plot(x2, y2, label='真实分布', color='orange')

    # 添加标题和标签
    
    ax.get_yaxis().set_visible(False)  # 隐藏y轴
    # 隐藏x轴
    ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False, labeltop=False)
    ax.spines['top'].set_visible(False)  # 隐藏上边框
    ax.spines['right'].set_visible(False)  # 隐藏右边框
    ax.spines['left'].set_visible(False)  # 隐藏左边框

    ax.text(0, stats.norm.pdf(0, loc=0, scale=1)+0.01, '$\mu_1$', fontsize=16, fontweight='bold', ha='center') 
    ax.text(3, stats.norm.pdf(3, loc=3, scale=1)+0.01, '$\mu_2$', fontsize=16, fontweight='bold', ha='center')
    # 显示图形
    #设置图例markdonw 
    ax.legend().set_frame_on(False)  # 隐藏图例边框
    ax.legend().set_loc((0.8, 0.8))  # 图例位置
    
    ax.vlines(x=0, ymin=0, ymax=stats.norm.pdf(0, loc=0, scale=1), color='dodgerblue', linestyle='--',  alpha=0.5)
    ax.vlines(x=3, ymin=0, ymax=stats.norm.pdf(3, loc=3, scale=1), color='orange', linestyle='--',  alpha=0.5)
    # 添加置信区间和显著性水平线

    ax.vlines(x=stats.norm.ppf(0.975, loc=0, scale=1), ymin=0, ymax=stats.norm.pdf(1.96, loc=3, scale=1), color='red', linestyle='--', label='$1-\\alpha/2$分位数')
    
    ax.fill_between(x1, y1, where=(x1 > stats.norm.ppf(0.975, loc=0, scale=1)), color='dodgerblue', alpha=0.2, label='$\\alpha/2$')
    ax.fill_between(x2, y2, where=(x2 < stats.norm.ppf(0.975, loc=0, scale=1)), color='orange', alpha=0.2, label='$\\beta$')
    
    # 填充 真实分布下方的区域，蓝色区域上方
    
    ax.fill_between(x2, stats.norm.pdf(x2,loc=3,scale=1), stats.norm.pdf(x2,loc=0,scale=1), where= (x2 > stats.norm.ppf(0.975, loc=0, scale=1)) & (x2 <3), color='green', alpha=0.2)
    ax.fill_between(x2, stats.norm.pdf(x2,loc=3,scale=1), where= (x2 > 3), color='green', alpha=0.2, label='$1-\\beta-\\alpha/2$')
   

    #填充两条线之间的区域
    
    plt.legend()
    plt.show()
```

```python
# 计算两样本均值检验的最小样本量，方差假设相同
import numpy as np
import scipy.stats as stats

def calculate_min_sample_size(alpha=0.05, beta=0.2, effect_size=0.5, sigma=1):
    """
    计算两样本均值检验的最小样本量
    :param alpha: 显著性水平
    :param beta: 第二类错误率
    :param effect_size: 效应大小
    :param sigma: 总体标准差
    :return: 每组的最小样本量
    """
    
    # z_alpha = stats.norm.isf(alpha / 2)  # 双尾检验的z值
    # z_beta = stats.norm.isf(1-beta)  # 第二类错误率的z值

    z_alpha = stats.norm.ppf(1- alpha / 2)  # 双尾检验的z值
    z_beta = stats.norm.ppf(1-beta)  # 第二类错误率的z值
    print(f'z_alpha: {z_alpha}, z_beta: {z_beta}')  # 打印z值
    print(effect_size, sigma, sigma **2)
    n = 2 * sigma **2 *(z_alpha + z_beta)**2  / (effect_size) ** 2  # 最小样本量公式
    return np.ceil(n).astype(int)  # 向上取整并转换为整数
```
