```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
# 绘制AB实验的学习效应示意图
def plot_learning_effect():
    # 数据
    x = np.arange(1, 11)  # 1到10的学习周期
    y_A = np.log(np.log(np.array(x))+1) # A组的学习效果

    # 绘图
    # 创建ax对象
     
    fig, axs = plt.subplots(1, 2, figsize=( 10,5), sharey=True,sharex=True)  # 创建1行2列的子图，设置共享y轴
    ax1 = axs[0]
    ax1.plot(x, y_A, marker='o',  color='orange', label= '整体')

    
    # 添加标题和标签
    # ax1.set_title('学习效应')
    ax1.set_xlabel('实验周期(周)')
    ax1.set_ylabel('实验效应')
    ax1.legend().set_frame_on(False)  # 隐藏图例边框
    ax1.legend().set_loc((0.9, 0.8))  #
    
    ax1.set_xlim(0, 11)  # 设置x轴范围
    # 添加网格和图例
    # ax1.legend().set_frame_on(False)  # 隐藏图例边框
    # ax1.legend().set_loc((0.9, 0.6))  # 图例位置
    
    ax1.spines['top'].set_visible(False)  # 隐藏上边框
    ax1.spines['right'].set_visible(False)  # 隐藏

    ax2 = axs[1]
    #绘制新老用户的实验效应
    y_B = np.log(np.log(np.array(x))+1)  # B组的学习效果
    # 绘制新用户效应 在0.4处上下波动
    y_new = 0.4 + 0.05 * np.random.randn(10)  # 新用户的实验效应
    ax2.plot(x, y_new, marker='o', label='新用户', color='gold')
    ax2.plot(x, y_B, marker='o', label='老用户', color='dodgerblue')
    # ax2.set_title('新老用户的实验效应')
    ax2.set_xlabel('实验周期(周)')
    ax2.legend().set_frame_on(False)  # 隐藏图例边框
    ax2.legend().set_loc((0.9, 0.75))  # 图例位置
    ax2.spines['top'].set_visible(False)  # 隐藏上边框
    ax2.spines['right'].set_visible(False)  # 隐藏右边框
    ax2.set_xlim(0, 11)  # 设置x轴范围
    # 显示图形
    plt.suptitle('学习效应示意', fontsize=16, fontweight='bold')  # 设置总标题
    plt.show()
plot_learning_effect()
```
