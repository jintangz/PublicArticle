```python
import os
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy 

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 指定默认
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
# 绘制正态分布曲线

x = np.linspace(-3, 5, 1000)
y = scipy.stats.norm.pdf(x, 1, 1)  # 标准正态分布
plt.plot(x, y, label='标准正态分布', color='black')
plt.xlim(-5, 7)
plt.ylim(0, 1)
frame = plt.gca()
frame.axes.get_xaxis().set_visible(False)
frame.axes.get_yaxis().set_visible(False)
frame.spines['top'].set_visible(False)  #隐藏上边框
frame.spines['right'].set_visible(False)  #隐藏右边框
frame.spines['left'].set_visible(False)  #隐藏左边框
frame.spines['bottom'].set_visible(False)

plt.vlines(-0.75, 0, 0.4, colors='gray', linestyles='dashed')
plt.text(-0.75,-0.04, '$z_{1-\\beta}$')
plt.fill_between(x, 0, y, where=(x <= -0.75), color='gray', alpha=0.5)
plt.text(2, 0.4, '$H_1: \\bar{X}_1-\\bar{X}_2\\sim{N(\\Delta_1, \\frac{\\sigma^2}{n_1}+\\frac{\\sigma^2}{n_2})}$', fontsize=12)
#隐藏x轴和y轴

# 嵌入另一个正态分布图 
inset_ax = plt.axes([0.1, 0.5, 0.4, 0.3])  # [left, bottom, width, height]
inset_x = np.linspace(-5, 5, 1000)
inset_y = scipy.stats.norm.pdf(inset_x, 0, 1)  # 均值为1，标准差为1的正态分布
inset_ax.plot(inset_x, inset_y, color='black')
inset_ax.set_xlim(-4, 4)
inset_ax.set_ylim(0, 0.5)
inset_ax.vlines(scipy.stats.norm.ppf(0.975), 0, 0.4, colors='gray', linestyles='dashed')
inset_ax.text(scipy.stats.norm.ppf(0.975),-0.04, '$z_{\\frac{\\alpha}{2}}$')
inset_ax.fill_between(inset_x, 0, inset_y, where=(inset_x >= scipy.stats.norm.ppf(0.975)), color='gray', alpha=0.5)
inset_ax.xaxis.set_visible(False)
inset_ax.yaxis.set_visible(False)
inset_ax.spines['top'].set_visible(False)
inset_ax.spines['right'].set_visible(False)
inset_ax.spines['left'].set_position(('data', 0))
inset_ax.spines['bottom'].set_position(('data', 0))
inset_ax.text(0, 0.45, '$H_0: \\bar{X}_1-\\bar{X}_2\\sim{N(0, \\frac{\\sigma^2}{n_1}+\\frac{\\sigma^2}{n_2})}$', fontsize=10)

plt.axis('off')  # 关闭坐标轴显示

plt.show()

```
